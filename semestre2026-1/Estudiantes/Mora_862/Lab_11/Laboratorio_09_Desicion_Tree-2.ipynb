
- Gini index


$G = 1 - (PC_1)^2 - (PC_2)^2$

$PC_1$: Probablidad categoria 1

$PC_2$: Probablidad categoria 2

Si :

- $G=0$ nodos puros, datos que pertenecen a una sola categoría

- Gini > 0 son nodos con impurezas, con mas de una categoría


La partición purpura  y amarilla del gráfico anterior representan nodos puros.

Mientras que la linea negra define:

- 7 puntos son de color azul.

- 6 puntos de color rojo.

$G = 1 - (7/13)^{2} - (6/13)^{2}$

mientrás que la ponderacion viene dada por:

$m = (7/20+7/20)=7/10$



No:

$G=1-(3/7)^{2}-(4/7)^{2}$
m = (3/20+4/20)


La funcion de coste, asocida al nodo padre:

$J(k, t_k) = \frac{m_{left}}{m} G_{left} + \frac{m_{right}}{m} G_{right} $

## SOLUCIÓN

````python
iris = load_iris()
X = iris.data
y = iris.target
````

### Dividir el conjunto de datos en conjuntos de entrenamiento y prueba

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
````

### Aplicar el algoritmo DecisionTreeClassifier

````python
clf = DecisionTreeClassifier(max_depth=4, random_state=0)
clf.fit(X_train, y_train)
````

### Analizar el gráfico de decisión

````python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 10), dpi=300)
tree.plot_tree(clf, class_names=iris.target_names, feature_names=iris.feature_names, impurity=False, filled=True, rounded=True)
plt.show()
````

### Características importantes

````python
caract = iris.data.shape[1]
plt.barh(range(caract), clf.feature_importances_)
plt.yticks(np.arange(caract),iris.feature_names)
plt.xlabel('Importancia de las características')
plt.ylabel('Características')
plt.show()
````

### Fronteras de desición

````python
# Parameters
n_classes = 3
plot_colors = "bry"
plot_step = 0.02

# Load data (already loaded, but keeping the original snippet for context)
# iris = load_iris()

plt.figure(figsize=(18, 12)) # Aumentar el tamaño general de la figura
plt.subplots_adjust(wspace=0.4, hspace=0.4) # Ajustar el espacio entre subplots

for pairidx, pair in enumerate([[0, 1], [0, 2], [0, 3],
                                [1, 2], [1, 3], [2, 3]]):
    # We only take the two corresponding features
    X_pair = iris.data[:, pair]
    y_pair = iris.target

    # Shuffle
    idx = np.arange(X_pair.shape[0])
    np.random.seed(13) # Set a random seed for reproducibility
    np.random.shuffle(idx)
    X_pair = X_pair[idx]
    y_pair = y_pair[idx]

    # Standardize
    mean = X_pair.mean(axis=0)
    std = X_pair.std(axis=0)
    X_pair = (X_pair - mean) / std

    # Train
    clf_pair = DecisionTreeClassifier().fit(X_pair, y_pair)

    # Plot the decision boundary
    plt.subplot(2, 3, pairidx + 1)

    x_min, x_max = X_pair[:, 0].min() - 1, X_pair[:, 0].max() + 1
    y_min, y_max = X_pair[:, 1].min() - 1, X_pair[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))

    Z = clf_pair.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)

    plt.xlabel(iris.feature_names[pair[0]])
    plt.ylabel(iris.feature_names[pair[1]])
    plt.axis("tight")

    # Plot the training points
    for i, color in zip(range(n_classes), plot_colors):
        idx = np.where(y_pair == i)
        plt.scatter(X_pair[idx, 0], X_pair[idx, 1], c=color, label=iris.target_names[i],
                    cmap=plt.cm.Paired)

    plt.axis("tight")

plt.suptitle("Decision surface of a decision tree using paired features", fontsize=16) # Aumentar el tamaño del título general
plt.legend(fontsize=10) # Ajustar el tamaño de la leyenda
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Ajustar el layout para evitar solapamiento, dejando espacio para el suptitle
plt.show()
````

Las gráficas de fronteras de decisión muestran cómo el árbol de decisión segmenta el espacio de características bidimensional para clasificar las especies de Iris. Cada subgráfico utiliza un par de características del dataset, con los puntos de datos reales coloreados según su especie. Las regiones coloreadas en el fondo de cada gráfico representan el área donde el modelo predice una clase específica, y las líneas rectas que las delimitan son las fronteras de decisión. Estas fronteras, al ser axiales, reflejan la naturaleza de un árbol de decisión que toma decisiones binarias basadas en umbrales de características individuales. La claridad y separación de estas regiones indican la eficacia del modelo para distinguir entre las clases utilizando ese par de características, siendo los pares que involucran petal length y petal width generalmente los más efectivos para lograr una distinción limpia entre las tres especies de Iris.
