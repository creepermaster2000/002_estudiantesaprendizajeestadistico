## Regresión lineal

Supongamos que tenemos un sistema en el que existe un predictor con **m** valores de entrenamiento, así:

$$(x^{(1)}, y^{(1)}), (x^{(2)}, y^{(2)}), \dots, (x^{(m)}, y^{(m)})$$

| Muestra | $Y$       | $X_1$       |
|--------:|:---------:|:-----------:|
| 1       | $Y^{(1)}$ | $X_1^{(1)}$ |
| 2       | $Y^{(2)}$ | $X_1^{(2)}$ |
| $\vdots$| $\vdots$  | $\vdots$    |
| $m$     | $Y^{(m)}$ | $X_1^{(m)}$ |

Podemos definir un modelo lineal como

$$
h(X)=\theta_0+\theta_1X
$$

donde $(\theta_0,\theta_1)$ son los parámetros del modelo. Nuestro objetivo es encontrar el conjunto de parámetros $(\theta_0,\theta_1)$ que se encuentre más "cercano" a $Y$ para cada valor de $X$.

Para la optimización, vamos a definir la función de coste **$J(\theta_0,\theta_1)$** para las muestras de entrenamiento como aquella que mide la distancia euclidiana respecto a la hipótesis planteada, así:

\begin{equation}
J(\theta_0,\theta_1)=\frac{1}{2m}\sum_{i=1}^{m}\left(h_{\theta}(x^{(i)})-y^{(i)}\right)^2
\end{equation}

Para encontrar los valores de $(\theta_0,\theta_1)$ se necesita minimizar la función de coste, lo que permite obtener los valores más cercanos. Esta minimización puede realizarse a través de diferentes métodos; el más conocido es el gradiente descendente.


![](https://github.com/hernansalinas/Curso_aprendizaje_estadistico/blob/main/Sesiones/imagenes/fig00.png?raw=true=50x)



Supongamos un modelo lineal para realizar la predicción. Así, nuestro modelo estará basado en la siguiente hipótesis de trabajo:

$$
h(X)=\theta_0+\theta_1X
$$

donde $\theta_0$ y $\theta_1$ corresponden a los parámetros del modelo.

Reemplazando el modelo anterior en la función de coste, obtenemos:

\begin{equation}
J(\theta_0,\theta_1)=\frac{1}{2m}\sum_{i=1}^{m}\left((\theta_0+\theta_1x^{(i)})-y^{(i)}\right)^2
\end{equation}

## Gradiente descendente

Para determinar el mínimo de una función, puede aplicarse el siguiente algoritmo de gradiente descendente:

- Proponer un número aleatorio inicial $\omega_i$.
- Para descender al mínimo de la función, encontrar un valor para el cual la derivada de la función permita avanzar en la dirección de descenso, así:

\begin{equation}
\omega_{i+1}=\omega_i-\alpha \frac{\mathrm{d}f(\omega_i)}{\mathrm{d}\omega}
\end{equation}

donde $\alpha$ es conocido como la tasa de aprendizaje del algoritmo.

- Evaluar $f(\omega_{i+1})$.
- Iterar hasta encontrar el mínimo de la función.

## Modelo de *machine learning*: solución general

Un modelo general para solucionar un problema de *machine learning* puede ser estructurado como sigue:

### a. Elegir el modelo a emplear

\begin{equation}
h(X,\Theta)
\end{equation}

- En el caso de una regresión lineal, tenemos que $h(X,\Theta=(\theta_0,\theta_1))$:

\begin{equation}
h(X)=\theta_0+\theta_1X
\end{equation}

### b. Elegir la función de coste

- Métrica euclidiana:

\begin{equation}
J(\Theta)=\frac{1}{2m}\sum_{i=1}^{m}\left(h_{\theta}(X^{(i)})-y^{(i)}\right)^2
\end{equation}

- [Lista de funciones de coste que pueden ser empleadas](https://jmlb.github.io/flashcards/2018/04/21/list_cost_functions_fo_neuralnets/)

### c. Aplicar el gradiente descendente iterativamente hasta encontrar el mínimo

\begin{equation}
\Delta \vec{\Theta}=-\alpha \nabla J(\Theta)
\end{equation}

- En el caso de una regresión lineal, tenemos que $h(X,\Theta=(\theta_0,\theta_1))$:

\begin{equation}
\theta_0 := \theta_0-\alpha \frac{\partial J}{\partial \theta_0}
\end{equation}

\begin{equation}
\theta_1 := \theta_1-\alpha \frac{\partial J}{\partial \theta_1}
\end{equation}

### Hint

Pueden comprobar la solución de la superficie a partir de la forma matricial de la función de coste:

\begin{equation}
J(\Theta)=\frac{1}{2m}\sum_{i=1}^{m}\left(h_{\Theta}(X^{(i)})-y^{(i)}\right)^2
\end{equation}

Sea

\begin{equation}
\Theta^T=[\theta_0,\theta_1]
\end{equation}

y sea la matriz de diseño

\begin{equation}
X=
\begin{bmatrix}
1 & 1 & 1 & \cdots & 1\\
x_1^{(1)} & x_1^{(2)} & x_1^{(3)} & \cdots & x_1^{(m)}
\end{bmatrix}
\end{equation}

de dimensión $(n+1)\times m$. En este caso, como solo se tiene una característica, $n=1$.

Luego, las predicciones del modelo pueden escribirse como

\begin{equation}
\Lambda=\Theta^T X=
\begin{bmatrix}
\theta_0+\theta_1 x_1^{(1)} &
\theta_0+\theta_1 x_1^{(2)} &
\theta_0+\theta_1 x_1^{(3)} &
\cdots &
\theta_0+\theta_1 x_1^{(m)}
\end{bmatrix}
\end{equation}

De esta manera, la función de coste se obtiene comparando estas predicciones con los valores reales $Y$:

\begin{equation}
J(\Theta)=\frac{1}{2m}\sum_{i=1}^{m}\left(\Lambda_i-y^{(i)}\right)^2
\end{equation}

o, de forma equivalente, si definimos

\begin{equation}
Y=
\begin{bmatrix}
y^{(1)} & y^{(2)} & y^{(3)} & \cdots & y^{(m)}
\end{bmatrix},
\end{equation}

entonces

\begin{equation}
J(\Theta)=\frac{1}{2m}\sum_{i=1}^{m}\left((\Theta^T X)_i-Y_i\right)^2
\end{equation}

Por tanto, en Python, la verificación numérica puede realizarse calculando primero las predicciones, luego el error y finalmente el promedio del error cuadrático.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

np.random.seed(42)
````

## Punto 1: DataFrame con datos lineales simples

````python
# Datos lineales: y = 2 + 3*x (sin ruido para caso ideal)
X_ideal = np.linspace(0, 10, 50)
y_ideal = 2 + 3 * X_ideal

df_ideal = pd.DataFrame({"X": X_ideal, "y": y_ideal})
print(df_ideal.head(10))
print(f"\nDimensiones: {df_ideal.shape}")

plt.figure(figsize=(8, 5))
plt.scatter(df_ideal["X"], df_ideal["y"], color="steelblue", label="Datos ideales")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Datos lineales simples (sin ruido)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

## Punto 2: Función de coste cuadrática

````python
def hypothesis(X, theta0, theta1):
    """Modelo lineal: h(X) = theta0 + theta1 * X"""
    return theta0 + theta1 * X


def cost_function(X, y, theta0, theta1):
    """
    Función de coste cuadrática (MSE/2):
    J(θ₀, θ₁) = (1/2m) * Σ(h(x⁽ⁱ⁾) - y⁽ⁱ⁾)²
    """
    m = len(y)
    predictions = hypothesis(X, theta0, theta1)
    J = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
    return J


# Verificación rápida con los datos ideales (θ₀=2, θ₁=3 debe dar J=0)
print(f"J(θ₀=2, θ₁=3) = {cost_function(X_ideal, y_ideal, 2, 3):.6f}")  # ≈ 0
print(f"J(θ₀=0, θ₁=3) = {cost_function(X_ideal, y_ideal, 0, 3):.6f}")  # > 0
print(f"J(θ₀=2, θ₁=0) = {cost_function(X_ideal, y_ideal, 2, 0):.6f}")  # > 0
````

## Punto 3: Función de coste con θ₀=0 fijo, variando θ₁

````python
# Fijar θ₀ = 0 y variar θ₁
theta1_values = np.linspace(-2, 8, 200)
J_values = [cost_function(X_ideal, y_ideal, 0, t1) for t1 in theta1_values]

# Encontrar el mínimo
min_idx = np.argmin(J_values)
theta1_opt = theta1_values[min_idx]
J_min = J_values[min_idx]
print(f"Mínimo de J con θ₀=0: θ₁* = {theta1_opt:.4f}, J = {J_min:.4f}")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Gráfica de J vs θ₁
axes[0].plot(theta1_values, J_values, "b-", linewidth=2)
axes[0].axvline(theta1_opt, color="r", linestyle="--", label=f"θ₁* = {theta1_opt:.3f}")
axes[0].scatter([theta1_opt], [J_min], color="red", s=100, zorder=5)
axes[0].set_xlabel("θ₁")
axes[0].set_ylabel("J(0, θ₁)")
axes[0].set_title("Función de coste con θ₀ = 0")
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Recta obtenida sobre los datos
axes[1].scatter(X_ideal, y_ideal, color="steelblue", label="Datos", alpha=0.6)
axes[1].plot(X_ideal, hypothesis(X_ideal, 0, theta1_opt), "r-", linewidth=2,
             label=f"h(X) = {theta1_opt:.3f}·X")
axes[1].set_xlabel("X")
axes[1].set_ylabel("y")
axes[1].set_title("Recta con θ₀=0 que minimiza J")
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

## Punto 4: Superficie y curvas de nivel con θ₀ y θ₁ variables

````python
# Meshgrid para θ₀ y θ₁
theta0_range = np.linspace(-5, 10, 100)
theta1_range = np.linspace(-2, 8, 100)
T0, T1 = np.meshgrid(theta0_range, theta1_range)

# Evaluar J en cada punto de la malla
J_surface = np.zeros_like(T0)
for i in range(T0.shape[0]):
    for j in range(T0.shape[1]):
        J_surface[i, j] = cost_function(X_ideal, y_ideal, T0[i, j], T1[i, j])

# Encontrar mínimo global
min_pos = np.unravel_index(np.argmin(J_surface), J_surface.shape)
t0_min = T0[min_pos]
t1_min = T1[min_pos]
print(f"Mínimo global: θ₀* ≈ {t0_min:.3f}, θ₁* ≈ {t1_min:.3f}, J = {J_surface[min_pos]:.6f}")

fig = plt.figure(figsize=(16, 6))

# Superficie 3D
ax1 = fig.add_subplot(121, projection="3d")
ax1.plot_surface(T0, T1, J_surface, cmap="viridis", alpha=0.8, edgecolor="none")
ax1.scatter([t0_min], [t1_min], [J_surface[min_pos]], color="red", s=100, zorder=5)
ax1.set_xlabel("θ₀")
ax1.set_ylabel("θ₁")
ax1.set_zlabel("J(θ₀, θ₁)")
ax1.set_title("Superficie de la función de coste")
ax1.view_init(elev=25, azim=45)

# Curvas de nivel
ax2 = fig.add_subplot(122)
contour = ax2.contour(T0, T1, J_surface, levels=30, cmap="viridis")
ax2.clabel(contour, inline=True, fontsize=7)
ax2.scatter([t0_min], [t1_min], color="red", s=100, zorder=5, label=f"Mínimo ({t0_min:.2f}, {t1_min:.2f})")
ax2.set_xlabel("θ₀")
ax2.set_ylabel("θ₁")
ax2.set_title("Curvas de nivel de J(θ₀, θ₁)")
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

## Punto 5: Interpretación geométrica

La función de coste $J(\theta_0, \theta_1)$ con la métrica euclidiana es un **paraboloide elíptico**. Su superficie tiene forma de "cuenco" con un único mínimo global.

- Las **curvas de nivel** son elipses concéntricas centradas en el mínimo $(\theta_0^*, \theta_1^*) = (2, 3)$, que son los parámetros verdaderos del modelo.
- La **elongación** de las elipses depende de la escala de los datos: como $X \in [0, 10]$, la función de coste es más sensible a cambios en $\theta_1$ que en $\theta_0$.
- El mínimo global es **único** (la función es estrictamente convexa) y corresponde al ajuste perfecto donde las predicciones coinciden exactamente con los datos.
- Cualquier desviación desde el mínimo en cualquier dirección incrementa $J$, lo cual garantiza la convergencia del gradiente descendente.

## Punto 6: Datos con ruido

````python
# Datos con ruido gaussiano
noise = np.random.normal(0, 3, size=len(X_ideal))
y_noisy = 2 + 3 * X_ideal + noise

df_noisy = pd.DataFrame({"X": X_ideal, "y": y_noisy})

# Superficie de coste con datos ruidosos
J_noisy = np.zeros_like(T0)
for i in range(T0.shape[0]):
    for j in range(T0.shape[1]):
        J_noisy[i, j] = cost_function(X_ideal, y_noisy, T0[i, j], T1[i, j])

min_pos_n = np.unravel_index(np.argmin(J_noisy), J_noisy.shape)
t0_min_n = T0[min_pos_n]
t1_min_n = T1[min_pos_n]
print(f"Caso ideal: θ₀* = 2.000, θ₁* = 3.000, J = 0.0")
print(f"Caso ruido: θ₀* ≈ {t0_min_n:.3f}, θ₁* ≈ {t1_min_n:.3f}, J = {J_noisy[min_pos_n]:.4f}")

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Datos
axes[0].scatter(X_ideal, y_ideal, alpha=0.5, label="Sin ruido", s=20)
axes[0].scatter(X_ideal, y_noisy, alpha=0.5, label="Con ruido", s=20)
axes[0].plot(X_ideal, hypothesis(X_ideal, t0_min_n, t1_min_n), "r-", linewidth=2,
             label=f"Ajuste ruido: {t0_min_n:.2f} + {t1_min_n:.2f}X")
axes[0].set_xlabel("X"); axes[0].set_ylabel("y")
axes[0].set_title("Comparación de datos")
axes[0].legend(); axes[0].grid(True, alpha=0.3)

# Contornos caso ideal
axes[1].contour(T0, T1, J_surface, levels=30, cmap="viridis")
axes[1].scatter([2], [3], color="red", s=100, marker="*", label="Mínimo ideal")
axes[1].set_xlabel("θ₀"); axes[1].set_ylabel("θ₁")
axes[1].set_title("Contornos - Sin ruido")
axes[1].legend(); axes[1].grid(True, alpha=0.3)

# Contornos caso ruidoso
axes[2].contour(T0, T1, J_noisy, levels=30, cmap="plasma")
axes[2].scatter([t0_min_n], [t1_min_n], color="red", s=100, marker="*", label=f"Mínimo ruido")
axes[2].scatter([2], [3], color="green", s=80, marker="x", label="Parámetros reales")
axes[2].set_xlabel("θ₀"); axes[2].set_ylabel("θ₁")
axes[2].set_title("Contornos - Con ruido")
axes[2].legend(); axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nComparación:")
print("  Sin ruido: el mínimo coincide exactamente con los parámetros reales.")
print("  Con ruido: el mínimo se desplaza ligeramente, pero J > 0 (ajuste imperfecto).")
print("  La forma del paraboloide se mantiene, pero el mínimo global ya no es J=0.")
````

## Punto 7: Efecto de un valor atípico (outlier)

````python
# Introducir un outlier extremo
X_outlier = np.append(X_ideal, 5.0)
y_outlier = np.append(y_ideal, 100.0)  # Punto (5, 100) vs valor real (5, 17)

# Calcular superficies con y sin outlier
J_outlier = np.zeros_like(T0)
for i in range(T0.shape[0]):
    for j in range(T0.shape[1]):
        J_outlier[i, j] = cost_function(X_outlier, y_outlier, T0[i, j], T1[i, j])

min_pos_o = np.unravel_index(np.argmin(J_outlier), J_outlier.shape)
t0_min_o = T0[min_pos_o]
t1_min_o = T1[min_pos_o]
print(f"Sin outlier: θ₀* = 2.000, θ₁* = 3.000")
print(f"Con outlier: θ₀* ≈ {t0_min_o:.3f}, θ₁* ≈ {t1_min_o:.3f}")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Datos con outlier
axes[0].scatter(X_ideal, y_ideal, alpha=0.6, label="Datos normales", s=30)
axes[0].scatter([5.0], [100.0], color="red", s=150, marker="X", label="Outlier (5, 100)", zorder=5)
axes[0].plot(X_ideal, hypothesis(X_ideal, 2, 3), "g--", linewidth=2, label="Ajuste sin outlier")
axes[0].plot(X_ideal, hypothesis(X_ideal, t0_min_o, t1_min_o), "r-", linewidth=2,
             label=f"Ajuste con outlier: {t0_min_o:.2f} + {t1_min_o:.2f}X")
axes[0].set_xlabel("X"); axes[0].set_ylabel("y")
axes[0].set_title("Efecto del outlier en la regresión")
axes[0].legend(); axes[0].grid(True, alpha=0.3)

# Contornos comparados
axes[1].contour(T0, T1, J_surface, levels=20, cmap="Greens", alpha=0.5)
axes[1].contour(T0, T1, J_outlier, levels=20, cmap="Reds", alpha=0.5)
axes[1].scatter([2], [3], color="green", s=100, marker="*", label="Mínimo sin outlier")
axes[1].scatter([t0_min_o], [t1_min_o], color="red", s=100, marker="*", label="Mínimo con outlier")
axes[1].set_xlabel("θ₀"); axes[1].set_ylabel("θ₁")
axes[1].set_title("Desplazamiento del mínimo por outlier")
axes[1].legend(); axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nDiscusión:")
print("  - La función de coste cuadrática penaliza el error al cuadrado, por lo que")
print("    un solo outlier puede desplazar significativamente el mínimo.")
print("  - Es sensible a valores extremos: un punto lejano 'arrastra' la recta hacia él.")
print("  - Limitación: no es robusta frente a outliers. Alternativas: MAE, Huber loss.")
````

## Punto 8: Modelo lineal sobre datos no lineales

## Punto 9: Expresión teórica de la función de coste

### Caso con un parámetro (θ₀ = 0)

Con $h(X) = \theta_1 X$:

$$J(\theta_1) = \frac{1}{2m}\sum_{i=1}^{m}(\theta_1 x^{(i)} - y^{(i)})^2 = \frac{\theta_1^2}{2m}\sum_i (x^{(i)})^2 - \frac{\theta_1}{m}\sum_i x^{(i)}y^{(i)} + \frac{1}{2m}\sum_i (y^{(i)})^2$$

Esta es una **parábola** en $\theta_1$. Su mínimo se obtiene derivando e igualando a cero:

$$\frac{dJ}{d\theta_1} = \frac{\theta_1}{m}\sum_i (x^{(i)})^2 - \frac{1}{m}\sum_i x^{(i)}y^{(i)} = 0$$

$$\theta_1^* = \frac{\sum_i x^{(i)}y^{(i)}}{\sum_i (x^{(i)})^2}$$

**Significado:** El mínimo es el valor de pendiente que mejor proyecta los datos sobre la dirección definida por $X$.

---

### Caso con dos parámetros

Con $h(X) = \theta_0 + \theta_1 X$:

$$J(\theta_0, \theta_1) = \frac{1}{2m}\sum_{i=1}^{m}(\theta_0 + \theta_1 x^{(i)} - y^{(i)})^2$$

Expandiendo:

$$J = \frac{1}{2m}\left[m\theta_0^2 + \theta_1^2\sum_i (x^{(i)})^2 + \sum_i(y^{(i)})^2 + 2\theta_0\theta_1\sum_i x^{(i)} - 2\theta_0\sum_i y^{(i)} - 2\theta_1\sum_i x^{(i)}y^{(i)}\right]$$

Es un **paraboloide elíptico** en el plano $(\theta_0, \theta_1)$. Las condiciones de mínimo (derivadas parciales = 0) dan el sistema:

$$\theta_0^* = \bar{y} - \theta_1^* \bar{x}$$

$$\theta_1^* = \frac{\sum_i (x^{(i)} - \bar{x})(y^{(i)} - \bar{y})}{\sum_i (x^{(i)} - \bar{x})^2}$$

**Significado:** $\theta_1^*$ es la covarianza entre $X$ e $Y$ dividida por la varianza de $X$, y $\theta_0^*$ asegura que la recta pasa por el centroide $(\bar{x}, \bar{y})$.

````python
# Verificación numérica de la solución analítica
# Caso 1 parámetro (θ₀=0):
theta1_analitico = np.sum(X_ideal * y_ideal) / np.sum(X_ideal**2)
print(f"Caso 1 parámetro (θ₀=0):")
print(f"  θ₁ analítico = {theta1_analitico:.6f}")
print(f"  θ₁ numérico  = {theta1_opt:.6f}")

# Caso 2 parámetros:
x_bar = np.mean(X_ideal)
y_bar = np.mean(y_ideal)
theta1_2p = np.sum((X_ideal - x_bar) * (y_ideal - y_bar)) / np.sum((X_ideal - x_bar)**2)
theta0_2p = y_bar - theta1_2p * x_bar
print(f"\nCaso 2 parámetros:")
print(f"  θ₀ analítico = {theta0_2p:.6f} (real = 2)")
print(f"  θ₁ analítico = {theta1_2p:.6f} (real = 3)")
````

## Punto 10: Gradiente descendente para f(x) = (x-4)²

````python
def gradient_descent_1d(f, df, x0, alpha, epsilon=1e-4, max_iter=10000):
    """
    Gradiente descendente en 1D.
    f: función objetivo
    df: derivada de f
    x0: punto inicial
    alpha: tasa de aprendizaje
    epsilon: tolerancia de convergencia
    """
    x = x0
    history = [x]
    for i in range(max_iter):
        grad = df(x)
        x_new = x - alpha * grad
        history.append(x_new)
        if abs(x_new - x) < epsilon:
            return x_new, history, i + 1
        x = x_new
    return x, history, max_iter


# f(x) = (x-4)², f'(x) = 2(x-4)
f = lambda x: (x - 4)**2
df = lambda x: 2 * (x - 4)

# Tres valores de α
alphas = [0.01, 0.1, 0.9]
x0 = 10.0  # Punto inicial alejado del mínimo

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
x_plot = np.linspace(-2, 12, 200)

for ax, alpha in zip(axes, alphas):
    x_min, history, n_iter = gradient_descent_1d(f, df, x0, alpha)
    history = np.array(history)

    ax.plot(x_plot, f(x_plot), "b-", linewidth=2, label="f(x) = (x-4)²")
    ax.plot(history, f(history), "ro-", markersize=4, alpha=0.6, label=f"GD (n={n_iter})")
    ax.scatter([history[-1]], [f(history[-1])], color="green", s=100, zorder=5,
               label=f"x* = {history[-1]:.4f}")
    ax.set_xlabel("x")
    ax.set_ylabel("f(x)")
    ax.set_title(f"α = {alpha}")
    ax.legend(fontsize=9)
    ax.grid(True, alpha=0.3)
    ax.set_ylim(-1, 40)

plt.suptitle("Gradiente descendente para f(x) = (x-4)² con ε = 10⁻⁴", fontsize=13)
plt.tight_layout()
plt.show()

print("\nResumen:")
for alpha in alphas:
    x_min, history, n_iter = gradient_descent_1d(f, df, x0, alpha)
    print(f"  α = {alpha}: x* = {x_min:.6f}, iteraciones = {n_iter}")
print("\n  α pequeño → convergencia lenta pero segura")
print("  α moderado → convergencia rápida")
print("  α grande → oscilaciones pero aún converge (si α < 1 para parábola)")
````

## Punto 11: Gradiente descendente para F(x,y) = sin(½x² - ¼y² + 3)·cos(2x + 1 - eʸ)

````python
# Definir F(x, y) y sus derivadas parciales (numéricas)
def F(x, y):
    return np.sin(0.5*x**2 - 0.25*y**2 + 3) * np.cos(2*x + 1 - np.exp(y))


def grad_F(x, y, h=1e-7):
    """Gradiente numérico por diferencias centrales"""
    dFdx = (F(x + h, y) - F(x - h, y)) / (2 * h)
    dFdy = (F(x, y + h) - F(x, y - h)) / (2 * h)
    return np.array([dFdx, dFdy])


# Superficie 3D
x_range = np.linspace(-3, 3, 200)
y_range = np.linspace(-2, 2, 200)
X_mesh, Y_mesh = np.meshgrid(x_range, y_range)
Z_mesh = F(X_mesh, Y_mesh)

fig = plt.figure(figsize=(16, 6))

# 3D surface
ax1 = fig.add_subplot(121, projection="3d")
ax1.plot_surface(X_mesh, Y_mesh, Z_mesh, cmap="coolwarm", alpha=0.8, edgecolor="none")
ax1.set_xlabel("x"); ax1.set_ylabel("y"); ax1.set_zlabel("F(x,y)")
ax1.set_title("Superficie de F(x,y)")
ax1.view_init(elev=30, azim=45)

# Contour map
ax2 = fig.add_subplot(122)
contour = ax2.contourf(X_mesh, Y_mesh, Z_mesh, levels=50, cmap="coolwarm")
plt.colorbar(contour, ax=ax2)
ax2.set_xlabel("x"); ax2.set_ylabel("y")
ax2.set_title("Mapa de contorno de F(x,y)")

plt.tight_layout()
plt.show()
````

````python
# Gradiente descendente 2D para encontrar el mínimo de F
def gradient_descent_2d(grad_func, x0, y0, alpha=0.01, epsilon=1e-4, max_iter=10000):
    """Gradiente descendente en 2D con gradiente numérico."""
    pos = np.array([x0, y0], dtype=float)
    history = [pos.copy()]
    for i in range(max_iter):
        grad = grad_func(pos[0], pos[1])
        pos_new = pos - alpha * grad
        history.append(pos_new.copy())
        if np.linalg.norm(pos_new - pos) < epsilon:
            return pos_new, np.array(history), i + 1
        pos = pos_new
    return pos, np.array(history), max_iter


# Probar desde varios puntos iniciales para explorar mínimos locales
starts = [(0.0, 0.0), (-1.0, 0.5), (1.5, -0.5), (-2.0, 1.0)]
alpha_2d = 0.01

fig, ax = plt.subplots(figsize=(10, 8))
contour = ax.contourf(X_mesh, Y_mesh, Z_mesh, levels=50, cmap="coolwarm", alpha=0.7)
plt.colorbar(contour, ax=ax)

best_val = np.inf
best_pos = None

colors = ["lime", "cyan", "yellow", "white"]
for (x0, y0), color in zip(starts, colors):
    pos_min, hist, n_iter = gradient_descent_2d(grad_F, x0, y0, alpha=alpha_2d)
    val = F(pos_min[0], pos_min[1])
    ax.plot(hist[:, 0], hist[:, 1], "o-", color=color, markersize=2, linewidth=1,
            alpha=0.8, label=f"({x0},{y0})→({pos_min[0]:.2f},{pos_min[1]:.2f}), F={val:.4f}, n={n_iter}")
    if val < best_val:
        best_val = val
        best_pos = pos_min

ax.scatter([best_pos[0]], [best_pos[1]], color="red", s=200, marker="*", zorder=10,
           label=f"Mejor mínimo: F={best_val:.4f}")
ax.set_xlabel("x"); ax.set_ylabel("y")
ax.set_title("Gradiente descendente sobre F(x,y) — múltiples inicios")
ax.legend(fontsize=8, loc="upper right")
plt.tight_layout()
plt.show()

print(f"\nMejor mínimo encontrado:")
print(f"  (x*, y*) = ({best_pos[0]:.6f}, {best_pos[1]:.6f})")
print(f"  F(x*, y*) = {best_val:.6f}")
````

## Punto 12: Regresión lineal con gradiente descendente

````python
# Datos del enunciado
X_reg = np.linspace(0, 1, 100)
y_reg = 0.2 + 0.2 * X_reg + 0.02 * np.random.random(100)


def linear_regression_gd(X, y, alpha=0.1, epsilon=1e-6, max_iter=50000):
    """
    Regresión lineal usando gradiente descendente.
    Retorna θ₀, θ₁ y el historial de coste.
    """
    m = len(y)
    theta0, theta1 = 0.0, 0.0
    cost_history = []

    for i in range(max_iter):
        predictions = theta0 + theta1 * X
        error = predictions - y

        # Gradientes
        d_theta0 = (1 / m) * np.sum(error)
        d_theta1 = (1 / m) * np.sum(error * X)

        # Actualización simultánea
        theta0_new = theta0 - alpha * d_theta0
        theta1_new = theta1 - alpha * d_theta1

        cost_history.append(cost_function(X, y, theta0, theta1))

        # Criterio de convergencia
        if abs(theta0_new - theta0) < epsilon and abs(theta1_new - theta1) < epsilon:
            theta0, theta1 = theta0_new, theta1_new
            break

        theta0, theta1 = theta0_new, theta1_new

    return theta0, theta1, cost_history


# Ejecutar gradiente descendente
theta0_gd, theta1_gd, cost_hist = linear_regression_gd(X_reg, y_reg, alpha=0.5)
print(f"Gradiente descendente:")
print(f"  θ₀ = {theta0_gd:.6f}")
print(f"  θ₁ = {theta1_gd:.6f}")
print(f"  Iteraciones: {len(cost_hist)}")
print(f"  J final: {cost_hist[-1]:.8f}")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Convergencia del coste
axes[0].plot(cost_hist, "b-", linewidth=1)
axes[0].set_xlabel("Iteración")
axes[0].set_ylabel("J(θ₀, θ₁)")
axes[0].set_title("Convergencia de la función de coste")
axes[0].set_yscale("log")
axes[0].grid(True, alpha=0.3)

# Ajuste sobre los datos
axes[1].scatter(X_reg, y_reg, alpha=0.5, s=20, label="Datos")
axes[1].plot(X_reg, hypothesis(X_reg, theta0_gd, theta1_gd), "r-", linewidth=2,
             label=f"GD: {theta0_gd:.4f} + {theta1_gd:.4f}X")
axes[1].set_xlabel("X"); axes[1].set_ylabel("y")
axes[1].set_title("Regresión lineal por gradiente descendente")
axes[1].legend(); axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````
