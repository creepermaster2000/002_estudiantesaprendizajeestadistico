# Lab 5: Regresión multivariada

Supongamos que tenemos un conjunto de características $X = X_1, X_2,\ldots,X_j,\ldots,X_n$ para realizar una predicción $y$ con valores esperados $\hat{y}$.

Cada $X_j$ puede escribirse como:

$X_1 = x_1^{(1)},x_1^{(2)},x_1^{(3)},\ldots,x_1^{(m)}$,

$X_2 = x_2^{(1)},x_2^{(2)},x_2^{(3)},\ldots,x_2^{(m)}$,

.

.

.

$X_n = x_n^{(1)},x_n^{(2)},x_n^{(3)},\ldots,x_n^{(m)}$.

Siendo $n$ el número de características y $m$ el número de datos,
$\hat{y} = \hat{y}^{(1)},\hat{y}^{(2)},\ldots,\hat{y}^{(m)}$ el conjunto de valores etiquetados, y
$y = y^{(1)},y^{(2)},\ldots,y^{(m)}$ los valores predichos por un modelo.

Lo anterior puede resumirse como:

| Training | $\hat{y}$ | $X_1$ | $X_2$ | . | . | . | . | $X_n$ |
|--------|--------|------|------|---|---|---|---|------|
| 1 | $\hat{y}^{(1)}$ | $x_1^{(1)}$ | $x_2^{(1)}$ | . | . | . | . | $x_n^{(1)}$ |
| 2 | $\hat{y}^{(2)}$ | $x_1^{(2)}$ | $x_2^{(2)}$ | . | . | . | . | $x_n^{(2)}$ |
| . | . | . | . | . | . | . | . | . |
| . | . | . | . | . | . | . | . | . |
| . | . | . | . | . | . | . | . | . |
| m | $\hat{y}^{(m)}$ | $x_1^{(m)}$ | $x_2^{(m)}$ | . | . | . | . | $x_n^{(m)}$ |

y el modelo puede ajustarse como sigue:

Para un solo dato de entrenamiento tenemos que:

\begin{equation}
y = h(\theta_0,\theta_1,\theta_2,\ldots,\theta_n) = \theta_0 + \theta_1 x_1+\theta_2 x_2 + \theta_3 x_3 + \ldots + \theta_n x_n
\end{equation}

\begin{equation}
h_{\Theta}(x) =
[\theta_0,\theta_1,\ldots,\theta_n]
\begin{bmatrix}
1\\
x_1\\
x_2\\
\vdots\\
x_n
\end{bmatrix}
= \Theta^T X
\end{equation}

Para todo el conjunto de datos, tenemos que:

Sea $\Theta^T = [\theta_0,\theta_1,\theta_2,\ldots,\theta_n]$ una matriz $1 \times (n+1)$ y

\begin{equation}
X =
\begin{bmatrix}
1 & 1 & 1 & \cdots & 1\\
x_1^{(1)} & x_1^{(2)} & x_1^{(3)} & \cdots & x_1^{(m)}\\
x_2^{(1)} & x_2^{(2)} & x_2^{(3)} & \cdots & x_2^{(m)}\\
\vdots & \vdots & \vdots & \ddots & \vdots\\
x_n^{(1)} & x_n^{(2)} & x_n^{(3)} & \cdots & x_n^{(m)}
\end{bmatrix}_{(n+1)\times m}
\end{equation}

luego,

\begin{equation}
h = \Theta^T X
\end{equation}

con dimensión $1\times m$.

La anterior ecuación es un hiperplano en $\mathbb{R}^n$. Nótese que, en caso de tener una sola característica, la ecuación puede analizarse según lo visto en la sesión de regresión lineal.

Para la optimización, vamos a definir la función de coste $J(\theta_0,\theta_1,\theta_2,\ldots,\theta_n)$ como la función asociada a la mínima distancia entre los valores predichos y los valores esperados, según la métrica euclidiana.

- Métrica euclidiana

\begin{equation}
J(\theta_0,\theta_1,\theta_2,\ldots,\theta_n)=\frac{1}{2m} \sum_{i=1}^m \left( h_{\Theta}(x^{(i)})-\hat{y}^{(i)}\right)^2
\end{equation}

Otras métricas pueden definirse como se muestra en la siguiente referencia: [Metricas](https://jmlb.github.io/flashcards/2018/04/21/list_cost_functions_fo_neuralnets/).

Nuestro objetivo será encontrar los valores
$\Theta = (\theta_0,\theta_1,\theta_2,\ldots,\theta_n)$
que minimizan el error respecto a los valores etiquetados y esperados $\hat{y}$.

Para encontrar el $\Theta$ óptimo, se necesita minimizar la función de coste, lo que permite obtener los valores más cercanos. Esta minimización puede realizarse a través de diferentes métodos; el más conocido es el gradiente descendente.


## Gradiente descendente

Consideremos la función de coste sin realizar el promedio  de funcion de coste:
\begin{equation}
\Lambda^T =
\begin{bmatrix}
(\theta_0 1 + \theta_1 x_1^1+\theta_2 x_2^2 + \theta_3 x_3^3 +...+ \theta_n x_n^n - \hat{y}^{1})^2 \\
(\theta_0 1+ \theta_1 x_1^1+\theta_2 x_2^2 + \theta_3 x_3^3 +...+ \theta_n x_n^n - \hat{y}^{2})^2\\
.\\
.\\
.\\
(\theta_0 1 + \theta_1 x_1^m+\theta_2 x_2^m + \theta_3 x_3^m +...+ \theta_n x_n^m - \hat{y}^{m})^2\\
\end{bmatrix}
\end{equation}

$\Lambda= [\Lambda_1,\Lambda_2, ...,\Lambda_m]$

$J = \frac{1}{2m} \sum_{i}^m \Lambda_i $

El gradiente descente, puede ser escrito como:

\begin{equation}
\Delta \vec{\Theta} =  - \alpha \nabla J(\theta_0, \theta_1,...,\theta_n)
\end{equation}

escogiendo el valor j-esimo tenemos que:

\begin{equation}
\theta_j :=  - \alpha \frac{\partial J(\theta_0, \theta_1,...\theta_j...,\theta_n)}{\partial \theta_j}
\end{equation}

Aplicando lo anterior a a función de coste asociada a la métrica ecuclidiana, tenemos que:

Para $j = 0$, 


\begin{equation}
\theta_0 :=  - \alpha \frac{\partial J(\theta_0, \theta_1,...\theta_j...,\theta_n)}{\partial \theta_0} = \frac{1}{m}\alpha \sum_{i=1}^m (\theta_j X_{ji} - \hat{y}^{(i)}) 1
\end{equation}



Para $0<j<n $

\begin{equation}
\theta_j :=  - \alpha \frac{\partial J(\theta_0, \theta_1,...\theta_j...,\theta_n)}{\partial \theta_j} = \frac{1}{m} \alpha\sum_{i=1}^m (\theta_{j} X_{ji} - \hat{y}^{(i)}) X_j
\end{equation}

donde X_j es el vector de entrenamiento j-esimo.

Lo  anterior puede ser generalizado como siguem, teniendo presente que $X_0 = \vec{1}$


Para $0\leq j<n$, 

\begin{equation}
\theta_j :=  - \alpha \frac{\partial J(\theta_0, \theta_1,...\theta_j...,\theta_n)}{\partial \theta_j} = \frac{1}{m} \alpha\sum_{i=1}^m (\theta_j X_{ji} - \hat{y}^{(i)}) X_j 
\end{equation}



# Vectorizando el grandiente descendete, tenemos que:
\begin{equation}
\nabla J = \Lambda^T X
\end{equation}

Luego:

\begin{equation}
\Theta=\Theta-\alpha \nabla J
\end{equation}

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plotly.graph_objects as go

np.random.seed(42)
````

## Punto 1: Generación de datos y visualización 3D

Definimos $y = 2.1 x_1 - 3.1 x_2$ y generamos puntos aleatorios sobre el plano.

````python
# Parámetros reales del plano
theta0_real = 0.0
theta1_real = 2.1
theta2_real = -3.1

# Generar puntos aleatorios
N = 100
x_1 = 2 * np.random.random(N) - 1  # x1 ∈ [-1, 1]
x_2 = 2 * np.random.random(N) - 1  # x2 ∈ [-1, 1]
y_label = theta1_real * x_1 + theta2_real * x_2  # sin ruido

print(f"Parámetros reales: θ₀={theta0_real}, θ₁={theta1_real}, θ₂={theta2_real}")
print(f"Puntos generados: {N}")
print(f"x₁ ∈ [{x_1.min():.3f}, {x_1.max():.3f}]")
print(f"x₂ ∈ [{x_2.min():.3f}, {x_2.max():.3f}]")

# Superficie del plano para referencia
x1_grid = np.linspace(-1, 1, 50)
x2_grid = np.linspace(-1, 1, 50)
X1_mesh, X2_mesh = np.meshgrid(x1_grid, x2_grid)
Y_mesh = theta1_real * X1_mesh + theta2_real * X2_mesh

# Gráfico 3D interactivo con Plotly
fig = go.Figure()
fig.add_trace(go.Surface(x=X1_mesh, y=X2_mesh, z=Y_mesh,
                         opacity=0.6, colorscale="Viridis",
                         name="Plano real"))
fig.add_trace(go.Scatter3d(x=x_1, y=x_2, z=y_label,
                           mode="markers",
                           marker=dict(size=4, color="red"),
                           name="Datos aleatorios"))
fig.update_layout(title="Plano y = 2.1·x₁ − 3.1·x₂",
                  scene=dict(xaxis_title="x₁", yaxis_title="x₂", zaxis_title="y"),
                  width=700, height=500)
fig.show()

# Gráfico 3D con matplotlib (alternativa estática)
fig_mpl = plt.figure(figsize=(10, 7))
ax = fig_mpl.add_subplot(111, projection="3d")
ax.plot_surface(X1_mesh, X2_mesh, Y_mesh, alpha=0.3, cmap="viridis")
ax.scatter(x_1, x_2, y_label, color="red", s=20, label="Datos aleatorios")
ax.set_xlabel("x₁"); ax.set_ylabel("x₂"); ax.set_zlabel("y")
ax.set_title("Plano y = 2.1·x₁ − 3.1·x₂")
ax.legend()
plt.tight_layout()
plt.show()
````

## Puntos 2-3: Inicialización de Θ y construcción de la matriz X

- Θ se inicializa aleatoriamente con dimensión $(n+1, 1) = (3, 1)$
- X tiene dimensión $(n+1, m) = (3, 100)$, con la primera fila de unos para el intercepto

````python
# Construir DataFrame
df = pd.DataFrame({"x1": x_1, "x2": x_2, "y": y_label})
df["ones"] = 1
print(df.head())

# Matriz X: (n+1, m) = (3, 100) — filas: [1, x1, x2]
X = df[["ones", "x1", "x2"]].values.T
Y = df["y"].values.reshape(1, N)

Nfeatures = X.shape[0]  # n+1 = 3
m = X.shape[1]           # 100
print(f"\nDimensiones:")
print(f"  X: {X.shape}  →  (n+1, m) = ({Nfeatures}, {m})")
print(f"  Y: {Y.shape}  →  (1, m)")

# Inicializar Θ aleatoriamente: (n+1, 1) = (3, 1)
theta = np.random.random(Nfeatures).reshape(Nfeatures, 1)
print(f"  Θ: {theta.shape}  →  (n+1, 1)")
print(f"\nΘ inicial = {theta.ravel()}")
````

## Punto 4: Función de coste vectorizada

Siguiendo las dimensiones indicadas:
- $h = \Theta^T X$ → (1, m)
- $\Lambda = h - Y$ → (1, m)
- $J = \frac{1}{2m}\sum_i \Lambda_i^2$

````python
def model(theta, X):
    """Hipótesis: h = Θᵀ X  →  (1, m)"""
    return theta.T @ X


def costo(h, Y, m):
    """Función de coste: J = (1/2m) Σ(h - ŷ)²"""
    Lambda = h - Y           # (1, m)
    Lambda_sq = Lambda ** 2  # (1, m)
    J = (1 / (2 * m)) * np.sum(Lambda_sq)
    return J


# Verificar dimensiones y coste inicial
h = model(theta, X)
J_init = costo(h, Y, m)
print(f"h = Θᵀ X  →  shape = {h.shape}")
print(f"Λ = h - Y  →  shape = {(h - Y).shape}")
print(f"Coste inicial J = {J_init:.6f}")
````

## Punto 5: Gradiente descendente vectorizado

$\nabla J = (h - Y) \cdot X^T$ → shape (n+1, 1) al transponer

$\Theta_{k+1} = \Theta_k - \alpha \nabla J$

````python
def update_params(theta, h, Y, X, learning_rate):
    """
    Gradiente descendente vectorizado.
    grad = ((h - Y) · Xᵀ)ᵀ  →  (n+1, 1)
    """
    grad = np.dot((h - Y), X.T).T  # (n+1, 1)
    theta = theta - learning_rate * grad
    return theta


# Verificar dimensiones del gradiente
h_test = model(theta, X)
grad_test = np.dot((h_test - Y), X.T).T
print(f"Gradiente shape: {grad_test.shape}  (debe ser ({Nfeatures}, 1))")
````

## Punto 6: Iteración — Encontrar Θ que ajusta el plano

Iteramos el gradiente descendente hasta convergencia y verificamos que los parámetros recuperados coincidan con $\theta_0=0$, $\theta_1=2.1$, $\theta_2=-3.1$.

````python
# Re-inicializar theta
theta = np.random.random(Nfeatures).reshape(Nfeatures, 1)
learning_rate = 0.01
Niter = 500
J_history = np.zeros(Niter)

print(f"Θ inicial: {theta.ravel()}")
print(f"Learning rate: {learning_rate}")
print(f"Iteraciones: {Niter}\n")

for i in range(Niter):
    h = model(theta, X)
    J_history[i] = costo(h, Y, m)
    theta = update_params(theta, h, Y, X, learning_rate)

print("═" * 55)
print("  RESULTADO DEL GRADIENTE DESCENDENTE")
print("═" * 55)
print(f"  θ₀ = {theta[0, 0]:+.6f}   (real =  0.0)")
print(f"  θ₁ = {theta[1, 0]:+.6f}   (real = +2.1)")
print(f"  θ₂ = {theta[2, 0]:+.6f}   (real = -3.1)")
print(f"  J final = {J_history[-1]:.10f}")
print("═" * 55)

# Gráficas
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Convergencia del coste
axes[0].plot(J_history, "b-", linewidth=1.5)
axes[0].set_xlabel("Iteración")
axes[0].set_ylabel("J(Θ)")
axes[0].set_title("Convergencia de la función de coste")
axes[0].set_yscale("log")
axes[0].grid(True, alpha=0.3)

# Comparación plano real vs ajustado
Y_pred = model(theta, X)
axes[1].scatter(Y.ravel(), Y_pred.ravel(), alpha=0.5, s=20, label="Predicciones")
lims = [Y.min(), Y.max()]
axes[1].plot(lims, lims, "r--", linewidth=2, label="y_pred = y_real")
axes[1].set_xlabel("y real")
axes[1].set_ylabel("y predicho")
axes[1].set_title("Predicción vs Valor real")
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

````python
# Plano ajustado vs plano real en 3D
Y_fit_mesh = theta[0, 0] + theta[1, 0] * X1_mesh + theta[2, 0] * X2_mesh

fig = go.Figure()
fig.add_trace(go.Surface(x=X1_mesh, y=X2_mesh, z=Y_mesh,
                         opacity=0.4, colorscale="Viridis",
                         name="Plano real", showscale=False))
fig.add_trace(go.Surface(x=X1_mesh, y=X2_mesh, z=Y_fit_mesh,
                         opacity=0.4, colorscale="Reds",
                         name="Plano ajustado", showscale=False))
fig.add_trace(go.Scatter3d(x=x_1, y=x_2, z=y_label,
                           mode="markers",
                           marker=dict(size=3, color="blue"),
                           name="Datos"))
fig.update_layout(title="Comparación: Plano real (verde) vs Plano ajustado (rojo)",
                  scene=dict(xaxis_title="x₁", yaxis_title="x₂", zaxis_title="y"),
                  width=700, height=500)
fig.show()
````

### Prueba con datos ruidosos

Verificamos que el gradiente descendente funciona también cuando los datos tienen ruido gaussiano.

````python
# Datos con ruido
y_noisy = theta1_real * x_1 + theta2_real * x_2 + 0.2 * (2 * np.random.random(N) - 1)
Y_noisy = y_noisy.reshape(1, N)

# Re-entrenar
theta_n = np.random.random(Nfeatures).reshape(Nfeatures, 1)
J_noisy_hist = np.zeros(Niter)

for i in range(Niter):
    h_n = model(theta_n, X)
    J_noisy_hist[i] = costo(h_n, Y_noisy, m)
    theta_n = update_params(theta_n, h_n, Y_noisy, X, learning_rate)

print("Con ruido:")
print(f"  θ₀ = {theta_n[0, 0]:+.6f}   (real =  0.0)")
print(f"  θ₁ = {theta_n[1, 0]:+.6f}   (real = +2.1)")
print(f"  θ₂ = {theta_n[2, 0]:+.6f}   (real = -3.1)")
print(f"  J final = {J_noisy_hist[-1]:.6f}")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(J_history, "b-", label="Sin ruido")
axes[0].plot(J_noisy_hist, "r-", label="Con ruido")
axes[0].set_xlabel("Iteración"); axes[0].set_ylabel("J(Θ)")
axes[0].set_title("Convergencia: sin ruido vs con ruido")
axes[0].set_yscale("log"); axes[0].legend(); axes[0].grid(True, alpha=0.3)

Y_pred_n = model(theta_n, X)
axes[1].scatter(Y_noisy.ravel(), Y_pred_n.ravel(), alpha=0.5, s=20)
lims = [Y_noisy.min(), Y_noisy.max()]
axes[1].plot(lims, lims, "r--", linewidth=2)
axes[1].set_xlabel("y real (con ruido)"); axes[1].set_ylabel("y predicho")
axes[1].set_title("Predicción vs Valor real (con ruido)")
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nCon ruido el J final no llega a 0, pero los parámetros se aproximan a los reales.")
````

7. Reescribir su código como una clase (ver ayuda)

````python
class MultilinearRegresion:
    def __init__(self, X, Y):
        """
        X: matriz de diseño (n+1, m) — incluye fila de unos
        Y: valores etiquetados (1, m)
        """
        self.X = X
        self.Y = Y
        self.Nfeatures = X.shape[0]  # n+1
        self.m = X.shape[1]          # m
        self.theta = np.random.random(self.Nfeatures).reshape(self.Nfeatures, 1)
        self.J_history = []

    def model(self):
        """Hipótesis: h = Θᵀ X  →  (1, m)"""
        self.h = self.theta.T @ self.X

    def costo(self):
        """Función de coste: J = (1/2m) Σ(h - ŷ)²"""
        self.J = (1 / (2 * self.m)) * np.sum((self.h - self.Y) ** 2)

    def update_params(self, learning_rate):
        """Gradiente descendente vectorizado."""
        grad = np.dot((self.h - self.Y), self.X.T).T  # (n+1, 1)
        self.theta = self.theta - learning_rate * grad

    def fit(self, learning_rate=0.01, n_iter=500):
        """Entrenar el modelo iterando el gradiente descendente."""
        self.J_history = []
        for i in range(n_iter):
            self.model()
            self.costo()
            self.J_history.append(self.J)
            self.update_params(learning_rate)
        # Calcular coste final
        self.model()
        self.costo()
        self.J_history.append(self.J)
        return self.theta

    def predict(self, X_new):
        """Predecir con nuevos datos."""
        return self.theta.T @ X_new
````

````python
# Entrenar con la clase — datos sin ruido
reg = MultilinearRegresion(X, Y)
print(f"Θ inicial: {reg.theta.ravel()}")

reg.fit(learning_rate=0.01, n_iter=500)

print("\n" + "═" * 55)
print("  RESULTADO CON CLASE MultilinearRegresion (sin ruido)")
print("═" * 55)
print(f"  θ₀ = {reg.theta[0, 0]:+.6f}   (real =  0.0)")
print(f"  θ₁ = {reg.theta[1, 0]:+.6f}   (real = +2.1)")
print(f"  θ₂ = {reg.theta[2, 0]:+.6f}   (real = -3.1)")
print(f"  J final = {reg.J:.10f}")
print("═" * 55)

# Entrenar con la clase — datos con ruido
reg_n = MultilinearRegresion(X, Y_noisy)
reg_n.fit(learning_rate=0.01, n_iter=500)

print(f"\n  CON RUIDO:")
print(f"  θ₀ = {reg_n.theta[0, 0]:+.6f}")
print(f"  θ₁ = {reg_n.theta[1, 0]:+.6f}")
print(f"  θ₂ = {reg_n.theta[2, 0]:+.6f}")
print(f"  J final = {reg_n.J:.6f}")

# Gráficas comparativas
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(reg.J_history, "b-", label="Sin ruido")
axes[0].plot(reg_n.J_history, "r-", label="Con ruido")
axes[0].set_xlabel("Iteración"); axes[0].set_ylabel("J(Θ)")
axes[0].set_title("Convergencia (clase MultilinearRegresion)")
axes[0].set_yscale("log"); axes[0].legend(); axes[0].grid(True, alpha=0.3)

# Plano ajustado con la clase
Y_class_pred = reg.predict(X)
axes[1].scatter(Y.ravel(), Y_class_pred.ravel(), alpha=0.5, s=20, label="Clase")
lims = [Y.min(), Y.max()]
axes[1].plot(lims, lims, "r--", linewidth=2, label="Perfecto")
axes[1].set_xlabel("y real"); axes[1].set_ylabel("y predicho")
axes[1].set_title("Predicción con la clase")
axes[1].legend(); axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````
